```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
customer_df=pd.read_csv('Ecommerce Customers.csv')
customer_df.head()
```

```python
customer_df.describe()
```

```python
customer_df.info()
```

```python
y=customer_df['Yearly Amount Spent']
X=customer_df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
check = pd.DataFrame(X_test, columns = ['Avg. Session Length','Time on App','Time on Website','Length of Membership'])
```

```python
check["Actual"] = y_test
```

```python
check["Predicted"] = y_pred
```

```python
lm.score(X_test,y_test)
```
